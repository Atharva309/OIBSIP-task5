```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/Advertising.csv'
df = pd.read_csv(file_path)
```

```python
# Plotting the graph of Advertisemnets vs Sales
plt.figure(figsize=(10, 6))

# Scatter plot of TV vs Sales in blue
plt.scatter(df['TV'], df['Sales'], color='blue', alpha=0.7, label='TV')

# Scatter plot of Radio vs Sales in red
plt.scatter(df['Radio'], df['Sales'], color='red', alpha=0.7, label='Radio')

# Scatter plot of Newspaper vs Sales in green
plt.scatter(df['Newspaper'], df['Sales'], color='green', alpha=0.7, label='Newspaper')

plt.title('Advertising vs Sales')
plt.xlabel('Advertising Budget ($)')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Using neural network regression model

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import r2_score

# Load the dataset
data = pd.read_csv("/content/Advertising.csv")

# Select the input and output features
X = data[["TV", "Radio", "Newspaper"]]
y = data["Sales"]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Scale the input features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build a neural network regression model
model = Sequential()
model.add(Dense(64, input_dim=3, activation="relu"))
model.add(Dense(32, activation="relu"))
model.add(Dense(1, activation="linear"))
model.compile(loss="mean_squared_error", optimizer="adam")

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Calculate the R-squared score
r2 = r2_score(y_test, y_pred)
print("R-squared score:", r2)
```

```python
4# performing Test on the model

# Ask the user to input values for the features
TV = float(input("Enter the TV advertisement budget: "))
Radio = float(input("Enter the Radio advertisement budget: "))
Newspaper = float(input("Enter the Newspaper advertisement budget: "))

# Create a dataframe with the user inputs
user_inputs = pd.DataFrame({"TV": [TV], "Radio": [Radio], "Newspaper": [Newspaper]})

# Scale the user inputs using the same scaler used for training the model
user_inputs_scaled = scaler.transform(user_inputs)

# Use the trained model to make a prediction on the user inputs
predicted_price = model.predict(user_inputs_scaled)

# Print the predicted value for the user inputs
print("Predicted Sales value:", predicted_price[0])
```
